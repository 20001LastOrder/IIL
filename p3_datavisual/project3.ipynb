### Borrow count history for each university

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.cbook as cbook
```

```python
borrowhistory = pd.read_csv('borrow.csv')  
```

```python
print((borrowhistory))
```

```python
plt.figure(figsize=(15, 15))
x = np.arange(len(np.array(borrowhistory['iname'])))
plt.bar(x, np.array(borrowhistory['borrowcount']))

plt.xticks(x, np.array(borrowhistory['iname']), rotation='vertical')
plt.xlabel('uni name')
plt.ylabel('borrow count ')
plt.show()
```
